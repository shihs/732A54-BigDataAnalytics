```python
# simulate data
# https://medium.com/@jacky308082/machine-learning--logistic-regression--python-d19b971ff9dc
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

np.random.seed(12)
number_observation=5000
#, list,x1,x2
x1=np.random.multivariate_normal([0,0],[[1,0.75],[0.75,1]],number_observation)
x2=np.random.multivariate_normal([1,4],[[1,0.75],[0.75,1]],number_observation)
#shape(5000,2) (10000,2) 
simulated_separableish_features=np.vstack((x1,x2)).astype(np.float32)

#colorlabel
simulated_labels=np.hstack((np.zeros(number_observation),np.ones(number_observation)))
plt.figure(figsize=(12,8))
plt.scatter(simulated_separableish_features[:,0],simulated_separableish_features[:,1],c=simulated_labels)
```

```python
print(simulated_separableish_features)
# print (simulated_separableish_features.shape)
simulated_labels = np.asarray(simulated_labels, dtype = int)
# simulated_labels = np.asarray(simulated_labels, dtype = str)
print (simulated_labels)
# print (simulated_labels.shape)

# simulated_labels = np.split(simulated_labels, 1, axis=1)
# print (simulated_labels)
# print(len(simulated_separableish_features))
# print (len(simulated_labels))

simulate = np.column_stack((simulated_labels, simulated_separableish_features))
np.savetxt("logistic_regression.csv", simulate, delimiter=",")
```
